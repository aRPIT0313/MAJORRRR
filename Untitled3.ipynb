```python
import requests
from bs4 import BeautifulSoup
import time

# Function to get the rest of the stock details (only once)
def get_stock_details(ticker):
    price_url = f'https://www.screener.in/company/{ticker}/consolidated/'

    # Fetch data from price_url
    response = requests.get(price_url)
    soup = BeautifulSoup(response.text, 'html.parser')

    class_name = 'margin-0 show-from-tablet-landscape'
    comp_name = soup.find(class_=class_name).text.strip()

    # Extract other details only once
    li_elements = soup.find_all('li', class_='flex flex-space-between')
    lb = [li.find_all('span', class_='number')[0].text.strip() for li in li_elements]

    # Store the details
    stock_details = {
        "comp_name": comp_name,
        "market_cap": lb[0],
        "high": lb[1],
        "low": lb[2],
        "stock_p_e": lb[3],
        "book_value": lb[4],
        "dividend_yield": lb[5],
        "roce": lb[6],
        "roe": lb[7],
        "face_value": lb[8]
    }

    return stock_details

# Function to get the real-time current price
def get_current_price(ticker):
    price_url = f'https://www.screener.in/company/{ticker}/consolidated/'

    # Fetch data from price_url
    response = requests.get(price_url)
    soup = BeautifulSoup(response.text, 'html.parser')

    class_current_price = 'flex flex-align-center'
    current_price_element = soup.find(class_=class_current_price).text.strip()
    price_with_symbol = current_price_element.split()[1]
    current_price = price_with_symbol.strip()

    return current_price

# Example ticker symbol to fetch data
ticker = 'TCS'

# Fetch the rest of the stock details once
stock_details = get_stock_details(ticker)

# Fetch and print other stock details once
print('NAME:', stock_details["comp_name"])
print('MARKET CAP:', stock_details["market_cap"])
print('HIGH:', stock_details["high"])
print('LOW:', stock_details["low"])
print('STOCK P/E :', stock_details["stock_p_e"])
print('BOOK VALUE :', stock_details["book_value"])
print('DIVIDEND YIELD :', stock_details["dividend_yield"])
print('ROCE : %', stock_details["roce"])
print('ROE : %', stock_details["roe"])
print('FACE VALUE:', stock_details["face_value"])
print('')

# Now, fetch and print the current price every 2 seconds for a set number of iterations
iterations = 100
for i in range(iterations):
    current_price = get_current_price(ticker)
    print(f"Iteration {i+1} - Current Price: {current_price}")
    time.sleep(0.2)  # Wait for 2 seconds before fetching again
```
